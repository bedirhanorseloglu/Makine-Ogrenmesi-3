Random Forest'ı ele almadan önce onun temelini oluşturan "bagging" yaklaşımını bir metodoloji olarak ele alacağız ve ondan sonra Random forest'a ilerlemiş olacağız.

# Bagging (Bootstrap Aggregating)

Temeli bootstrap (yerine koymalı seçme) yöntemi ile oluşturulan birden fazla karar ağacının ürettiği tahminlerin bir araya getirilerek değerlendirilmesine dayanır.
Bagging, aynı veri setinin farklı alt örnekleri üzerinde modellenen birden fazla modelin sonucunu birleştirerek tahminlerimizin varyansını azaltmak için kullanılan bir topluluk tekniğidir.



# Peki ne demek bu Bootstrap Rastgele Örnekleme  Yöntemi?

    Elimizde bir veri seti vardı. Bu veri seti üzerine bir tane makine öğrenmesi algoritması import ettik. Fit ettik. Şimdi bu yaklaşımın mimarı olan Breiman diyor ki "Benim yapmış olduğum CART algoritmasını ben bir veri setine uyguladığımda bu veri seti üzerinden tahminler elde ettiğimde görüyorum ki bu bazı aşırı öğrenme (overfitting) problemlerine yol açıyor. Bu noktada ya veriyi beslemem lazım ya da farklı bir çözüm getirmem lazım." İlk getirmiş olduğu çözüm de "Budama Çözümü" oluyor. Yani ağaçların dallanmalara ayrılmalarını öyle bir noktada kesmeliyim ki rasthgelelik, genellenebilirlik kaygısı bozulmasın. Fakat bu çabalarına rağmen yine de yeteri kadar aşırı öğrenmenin önüne geçemediğinden yine kendisinin ürettiği "Bagging" yöntemi ile ilgili bir geliştirme yapmıştır.

    Bagging yönteminde ise yine bir veri setimiz var. Bu veri seti içerisinden yerine koyacak şekilde farklı sayılarda örneklemler çekiliyor. Yani aslında model kurmak için bir tane veri varken örneğin 10 farklı veri oluyor ve 10 farklı ağaç kurmuş oluyor. Budama yönteminde ise tek bir ağaç vardı. Ve bu oluşan 10 ağaç üzerinden modelleme çalışması yapıyor.
    

# Bagging Çalışma Prensibi:
    
Torbalama işleminde yer alan ana adımlar şunlardır:

- Birden fazla veri kümesi oluşturma: Orijinal veri kümesi üzerinde bir değiştirme ile örnekleme yapılır ve orijinal veri kümesinden yeni veri kümeleri oluşturulur.
- Birden fazla model oluşturma: Bu küçük veri kümelerinin her birinde bir sınıflandırıcı oluşturulur, genellikle tüm veri kümelerinde aynı model oluşturulur.
- Modellerin Birleştirilmesi: Tüm bireysel sınıflandırıcıların tahminleri, genellikle öncekine kıyasla çok daha az varyansla daha iyi bir model vermek için birleştirilir.

 Bagging, Böl ve fethet'e benzer. Daha iyi doğruluk ve model kararlılığı elde etmek için bir araya getirilen orijinal veri kümesinden birden fazla alt küme üzerinde çalıştırılan bir grup tahmin modelidir.

 Çalışma prensibinin kilit noktası, bootstrap rastgele örnekleme yöntemidir. Bootstrap rastgele örnekleme yöntemi, gözlem birimlerinin içinden yerine koymlaı bir şekilde tekrar tekrar örnek çekmek demektir. Bu çekilen örneklemlere de tek tek ağaçlar fit edilir. Ve böylece birçok ağaç elde edilmiş olunur.

Bagging Yöntemi:
    
    - Hata kareler ortalamasının karekök değerini (RMSE) düşürür.
    - Doğru sınıflandırma oranını artırır.
    - Varyansı düşürür ve ezberlemeye karşı dayanıklıdır. 

# RANDOM FOREST

Temeli birden çok karar ağacının ürettiği tahminleri bir araya getirilerek değerlendirmesine dayanır.

Bagging yöntemini kullanır. Bagging'de kullandığı yöntemin üzerine ek bir yöntem ekler. Gözlemlerin rastgele seçilmesinin rassallığı artırdığını ve hataları düşürdüğünü gören Breiman bunun üzerine aynı işlemi değişken seçimi üzerinde de gerçekleştirmiştir. Breimann, gözlem (değişken) seçimine de bir rassallık getirerek çok çok daha başarılı bir performans elde edebilmiştir. Daha öncesinde CART'ta "splitter" hiperparametresini "splitter=best" olarak belirliyorduk. Bu aynen devam edecek ve ek olarak değişkenlerin de rastgele seçilmesini sağlayacağız..

Random Forest algoritması, Bagging ile Random Subspace yöntemlerinin birleşimi ile oluşturulmuştur.

Ağaçlar için gözlemler, bootstrap rastgele örnek seçim yöntemi ile ; değişkenler, random subspace yöntemi ile seçilir.

Karar ağacının her bir düğümünde en iyi dallara ayırıcı değişken tüm değişkenler, tüm değişkenler arasından rastgele seçilen daha az sayıdaki değişken arasından seçilir. (Örneğin elimizde p adet değişken olduğunda ondan daha küçük bir sayı olan r adet sayıda değişkeni rastgele seçeceğiz -> Bagging'de gözlemleri seçtiğimiz gibi)

Ağaç oluşturmada veri setinin 2/3'ü kullanılır. Dışarıda kalan veri, ağaçların performans değerlendirmesi ve değişken öneminin belirlenmesi için kullanılır.

Her düğüm noktasında rastgele değişken seçimi yapılır. (regresyonda -> p/3 tane ; sınıflandırmada -> karekök p tane değişken seçilir. ; (p adet değişken olduğunu yukarıdaki örnekte belirtmiştik.))

En önemli parametreleri : n_estimators , max_depth , min_samples_split , max_features(bölünmelerde göz önünde bulundurulması gereken değişken sayısını ifade eder.)
