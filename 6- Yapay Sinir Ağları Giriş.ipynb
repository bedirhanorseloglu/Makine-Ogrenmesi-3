İnsan beyninin bilgi işleme şeklini referans alan sınıflandırma ve regresyon problemleri için kullanılabilen kuvvetli makine öğrenmesi algoritmalarından birisidir.

Amacımız, en küçük hata ile tahmin yapabilecek katsayılara erişmektir.



Burada bilinmesi gereken şudur:

- Girdilerden gelen bilgiler belirli ağırlıklarla çarpıldıktan sonra toplanma fonksiyonuna aktarılır. Ve bu değerler aktivasyon fonksiyonunu tarafından işlenerek çıktı değerleri oluşturulur. 

- Regresoyon problemleri için sayısal değerler, sınıflandırma problemleri için de kategorik değişkenker (sınıflar) ifade edilmiş olacak.



Örnek yapay sinir ağı (Çok Katmanlı Yapay Sinir Ağı):

    - Bu ağlarda öğrenme kuralı olarak "delta öğrenme kuralı" kullanılır. Delta Öğrenme Kuralı'nda, kabul edilebilir bir hata miktarı elde edilinceye kadar ağ çıktısı ile gerçek çıktı arasındaki farklar minimize edilmeye çalışılır. Bu farkları minimize edebilmek adına girdilerin ağırlıkları değiştirilir. Bu ağırlık değiştirme işlemine teknik olarak "ağ öğreniyor" denilir.
