Support Vector Regression (SVR) güçlü ve esnek modelleme tekniklerinden birisidir.

Sınıflandırma ve regresyon için kullanilabilir.

Robust (dayanıklı - aykırı gözlemlere karşı) bir regresyon modelleme tekniğidir.

Amacımız, bir marjin aralığına maksimum noktayı en küçük hata ile alabilecek şekilde doğru ya da eğriyi belirlemektir.

Kısileri ve epsilon değerlerini kullanarak regresyon doğrusunun eğimini belirliyor olacağız.



Çözmemiz gerekek Minimizasyon Problemi ve Kısıtlamalar:



    - Doğrunun yönünü yani eğimini belirlemede epsilon değerinden küçük olan değerlerin bir etkisi yoktur. Fakat epsilon değerinin üstüne çıktığımızda (örneğin üst taraftaki aykırı değer) bu noktalar regresyon doğrusunu belirleyen noktaların ta kendisi. Yani hoşlanmadığımız artıklar SVR'de regresyon doğrusunu belirlemede kritik rol oynuyor. İşte bu sebeple biz tamamen kontrolü bunlara bırakmayalım, kendimiz de bir kontrol mekanizması buraya yerleştirelim diye formüldeki C ceza (kontrol , karışıklık) parametresi ekledik.
