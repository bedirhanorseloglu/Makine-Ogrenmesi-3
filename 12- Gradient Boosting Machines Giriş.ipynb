# Boosting Yöntemlerine Giriş

- Zayıf öğrencileri (random forest'taki kötü tahmin üreten ağaçlar) bir araya getirip güçlü bir öğrenci ortaya çıkarma fikrine dayanır.
- **AdaBoost (Adaptive Boosting):** Zayıf sınıflandırıcıların bir araya gelerek güçlü bir sınıflandırıcı oluşturması fikrini hayata geçiren algoritmadır. İlk çıktığında sınıflandırma problemleri için kullanılmıştır.

    Aşağıdaki görselde mavi içindeki maviler(+) ve kırmızı içindeki kırmızılar(-) doğru sınıflandırma örneğidir. 2. kutuda ilk aşamada yanlış sınıflandırılan örnekler göz önüne alınarak başka bir sınıflandırma yapılır. 3. kutuda tekrar yanlış sınıflandırılanlara odaklanıldığında üstteki bölümde bir eksi alttaki bölümde iki artı yanlış sınıflandırılmış olarak kaldı. Sonuç olarak 4. kutuda gözlemlendiği gibi tüm + ve -'ler doğru sınıflandırma içindedir.



- GBM, AdaBoost'un (Adaptive Boosting) sınıflandırma ve regresyon problemlerine kolayca uyarlanabilen genelleştirilmiş versiyonudur.
- GBM, artıklar(gerçek-tahmin) üzerine tek bir tahminsel model formunda olan modeller serisi kurulur.
- Seri içerisindeki bir model serideki bir önceki modelin tahmnin artıklarının(hatalarının) üzerine kurularak oluşturulur. (1. kutudaki hatanın üzerine 2. kutu , 2.dekinin üzerine 3. kutu ...)
- GBM, diferansiyellenebilen herhangi bir kayıt fonksiyonunu optimize edebilen Gradient Descent(optimizasyon algoritması) algoritmasını kullanır.
- GB, bir çok temel öğrenici tipi (base learner type) kullanabilir. (Trees, linear , terms , splines)
- Cost fonksiyonları ve link fonksiyonları modifiye edilebilirdir.
- GBM = Boosting + Gradient Descent (optimizasyon algoritması)
