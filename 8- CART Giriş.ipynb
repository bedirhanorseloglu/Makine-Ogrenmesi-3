Heterojen veri seti (verinin ilk işlenmemiş hali), belirlenmiş bir hedef (bağımlı ) değişkene göre homojen gruplara ayrılır.

Karar ağaçları, sınıflandırma ya da regresyon problemlerinde kullanılabilen gözetimli yöntemlerden birisidir.

Aşırı öğrenmeye (overfitting) eğilimli bir algoritmadır.

Büyük veri setlerindeki başarısı diğer algoritmalara göre biraz daha düşük kalmaktadır.

Bu bölümle birlikte ağaca dayalı makine öğrenmesi algoritmalarına girişi yapmış oluyoruz.

Adım:2

Tahmin işlemi yapıldıktan sonra hata durumumuzu kontrol ederiz. Hata durumumuzu gerçek değerden tahmin edilen değeri çıkardığımızda buluyoruz. Tablomuza göre satış değerlerinden tahmin değerlerimizi çıkaracağız.

# Sınıflandırma Problemleri için Cost Fonksiyonu:

Gini Katsayısı



Gini katsayısı dalın, kutunun ya da yaprağın saflığını belirler. Homojenliğin ölçüdür. Sınıflandırma problemlerinde elimizde 1 ve 0 sınıfı olmak üzere iki tane sınıf olur.

Formulü açıklayalım.

Adım1:

0 sınıfı için p(i), sıfır sınıfının gerçekleşme olasılığıdır. (1-p(i)) ise 0 sınıfının gerçekleşmeme olasılığıdır. Gerçekleme olasılığı ile gerçekleşmeme olasılığı çarpılır.

Adım:2

1 sınıfı için p(i), 1 sınıfının gerçekleşme olasılığıdır. (1-p(i)) ise 1 sınıfının gerçekleşmeme olasılığıdır. Gerçekleme olasılığı ile gerçekleşmeme olasılığı çarpılır.

Adım:3

(0 sınıfının gerçekleşme olasılığı * gerçekleşmeme olasılığı) + (1 sınıfının gerçekleşme olasılı * gerçekleşmeme olasılığı) = Gini katsayısıdır.

Gini katsayısına göre bölünmeye devam edilip edilmeyeceğine bakabiliriz. Gini katsayımız 0 olduğunda dalımız homojendir bölünmeye ihtiyaç yoktur deriz. Katsayımız küsüratlı bir sayı ise dalımız hala daha homojen değil, tekrar dallanmaya ihtiyaç vardır demiş oluyoruz.
